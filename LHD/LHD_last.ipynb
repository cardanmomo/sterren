```python
filename = 'rho65'

nkx0    = 64
nky0    = 96 
Lx      = 46.961
kxmin   = 0.1338
kymin   = 0.038

a  = 0.6361227
Lr = 1.65
R0  = 3.6731507

omt = 6

factor   = ((1/a)/(1/R0))**2
factor2   = ((1/Lr)/(1/R0))**2

fig = plt.figure(figsize=(13, 7))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.set_xlabel('$k_{y}  \\rho_{ti}$',fontsize=19)
ax1.set_ylabel('$\mathcal{S}_{\mathcal{T}} \ (k_{y} \\rho_{ti})$',fontsize=19)
ax2.set_xlabel('$k_{x}  \\rho_{ti}$',fontsize=19)
ax2.set_ylabel('$\mathcal{S}_{\mathcal{Z}} \ (k_{x} \\rho_{ti})$',fontsize=19)

top  = np.int(nkx0/2)
ky   = np.zeros((top, nky0))
phi  = np.zeros((top, nky0))
St   = np.zeros((nky0))
St1   = np.zeros((nky0))
St2   = np.zeros((nky0))

spec = [( pylab.loadtxt('{0}/ITGa_NL/omt{1}/spectraions_act.dat'.format(filename, omt),usecols = (0, 1)))][0]

for n in range(0, top):  # Extract ky and Phi for each nkx0
    lo     = (0+(nky0*n))
    hi     = (nky0+(nky0*n))
    ky[n]  = spec[lo:hi, 0]
    phi[n] = spec[lo:hi, 1]
    np.add(St, phi[n]/kymin, out=St, where=True)  # Complete range with normalisation
    np.add(St1, phi[n]*factor/kymin, out=St1, where=True)  # Complete range with normalisation
    np.add(St2, phi[n]*factor2/kymin, out=St2)  # Complete range with normalisation

hi = nky0*top
kx = spec[hi:(hi+top), 0]
Sz = spec[hi:(hi+top), 1]/kxmin
Sz1 = spec[hi:(hi+top), 1]*factor/kxmin
Sz2 = spec[hi:(hi+top), 1]*factor2/kxmin

idx = (np.abs(ky[1, :] - 0.07)).argmin()  # Truncate the integrals taking Nunami's range
idx2 = (np.abs(ky[1, :] - 0.4)).argmin()
Tn = trapz(St[idx:idx2], x=ky[1, idx:idx2])

T = trapz(St[1:], x=ky[1, 1:])
Z = trapz(Sz, x=kx)

ax1.loglog(ky[1, 1:], St[1:], linewidth=2)  # Printing: the truncated range with normalisation
ax1.loglog(ky[1, 1:], St1[1:], linewidth=2) 
ax1.loglog(ky[1, 1:], St2[1:], linewidth=2) 
ax2.loglog(kx, Sz, linewidth=2)
ax2.loglog(kx, Sz1, linewidth=2)
ax2.loglog(kx, Sz2, linewidth=2)
    
ax1.legend(['no factor', 'a', 'Lref'], fontsize=10)
plt.show()
```

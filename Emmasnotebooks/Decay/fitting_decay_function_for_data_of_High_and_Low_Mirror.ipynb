```python
from scipy.optimize import curve_fit

def func(x,a,b):
    return np.cos(a*x)*np.exp(-b*x)
```

```python
xdata = t17
ydata = E17
plt.plot(xdata, ydata, 'b-', label='data')

popt, pcov = curve_fit(func, xdata, ydata)
popt
plt.plot(xdata, func(xdata, *popt), 'r--',label='fit: a=%f, b=%f' % tuple(popt),linewidth=3)

#plt.title('W7-X High Mirror, s=0.36, $L_{x}=104.719$', fontsize=30)
plt.xlabel('Time ($a / v_{ti}$)',fontsize=20)
plt.ylabel('Re[$\phi (t) / \phi (0)]$',fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.axhline(y=0,linestyle='dashed',color='black')
plt.xlim([0,3000])
plt.tight_layout(rect=[0, 0, 3, 2])
plt.legend(borderpad=1.5,labelspacing=1.5,fontsize=16)
plt.show()
```

```python
from scipy.optimize import curve_fit

def func(x,a,b):
    return np.cos(a*x)*np.exp(-b*x)

xdata = t1
ydata = E1
plt.plot(xdata, ydata, 'b-', label='data')

popt, pcov = curve_fit(func, xdata, ydata)
popt
plt.plot(xdata, func(xdata, *popt), 'r--',label='fit: a=%5.3f, b=%5.3f' % tuple(popt),linewidth=3)

plt.title('W7-X High Iota, s=0.5, $L_{x}=104.719$', fontsize=30)
plt.xlabel('Time ($a / v_{ti}$)',fontsize=30)
plt.ylabel('Re[$\phi (t) / \phi (0)]$',fontsize=30)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.tight_layout(rect=[0, 0, 3, 2])
plt.legend(borderpad=1.5,labelspacing=1.5,fontsize=16)
plt.show()
```

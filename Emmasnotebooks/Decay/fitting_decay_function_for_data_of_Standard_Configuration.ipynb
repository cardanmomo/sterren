```python
from scipy.optimize import curve_fit

def func(x,a,b):
    return np.cos(a*x)*np.exp(-b*x)
```

```python
xdata = t7
ydata = E7
avalues = []
bvalues = []
plt.plot(xdata, ydata, 'b-', label='data')

popt, pcov = curve_fit(func, xdata, ydata)
print(popt)
print('avalues before are', avalues)
print('bvalues before are', bvalues)

plt.plot(xdata, func(xdata, *popt), 'r--',label='fit: a=%f, b=%f' % tuple(popt),linewidth=3)
plt.title('W7-X Standard Configuration, s=0.25, $L_{x}=104.719$', fontsize=25)
plt.xlabel('Time ($a / v_{ti}$)',fontsize=30)
plt.ylabel('Re[$\phi (t) / \phi (0)]$',fontsize=30)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.tight_layout(rect=[0, 0, 3, 2])
plt.legend(borderpad=1.5,labelspacing=1.5,fontsize=16)
plt.show()

avalues.append(popt[0])
bvalues.append(popt[1])
print('avalues after are', avalues)
print('bvalues after are',bvalues)
```

```python
data28 = pylab.loadtxt('C:/Users/s162605/Documents/Studie/JAAR3/BEP/sterren/IFERC/W7Xlms05/GAM/104.719/gamdiagions_act.dat')
E28 = (data28[:,1]*(1/data28[0, 1]))
e28 = (data28[:,2]*(1/data28[0, 1]))
t28 = data28[:,0]

data29 = pylab.loadtxt('C:/Users/s162605/Documents/Studie/JAAR3/BEP/sterren/IFERC/W7Xlms05/GAM/136.837/gamdiagions_act.dat')
E29 = (data29[:,1]*(1/data29[0, 1]))
e29 = (data29[:,2]*(1/data29[0, 1]))
t29 = data29[:,0]

Q28 = np.sqrt(E28**2+e28**2)
Q29 = np.sqrt(E29**2+e29**2)
```

```python
fig, ax1 = plt.subplots(figsize=(13, 7))
plt.title('W7-X Low Mirror, s:0.5', fontsize=30)
ax1.set_xlabel('Time ($a / v_{ti}$)',fontsize=16)
ax1.set_ylabel('$\phi (t) / \phi (0)$',fontsize=16)
ax1.axhline(y=0, color='black', linestyle='dashed')
plt.xlim([-10,500])

ax1.plot(t28, e28, linewidth=2, label=('Im W7Xlms05 104.719'),color='blue',linestyle='dotted')
ax1.plot(t28, E28, linewidth=2, label=('Re W7Xlms05 104.719'),color='blue',linestyle='dashed')
ax1.plot(t28, Q28, linewidth=2, label=('Mod W7Xlms05 104.719'),color='blue')
ax1.plot(t29, e29, linewidth=2, label=('Im W7Xlms05 136.837'),color='purple',linestyle='dotted')
ax1.plot(t29, E29, linewidth=2, label=('Re W7Xlms05 136.837'),color='purple',linestyle='dashed')
ax1.plot(t29, Q29, linewidth=2, label=('Mod W7Xlms05 136.837'),color='purple')

plt.legend(loc='upper center', 
             bbox_to_anchor=(0.5, -0.2),fancybox=False, shadow=False, ncol=2)
fig.tight_layout()
#plt.savefig('../../CMSS_2018/Poster/Salpha_nlzf_plot.pdf', format='pdf', dpi=1200)    
```

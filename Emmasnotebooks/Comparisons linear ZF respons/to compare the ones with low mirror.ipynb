```python
data28 = pylab.loadtxt('C:/Users/s162605/Documents/Studie/JAAR3/BEP/sterren/IFERC/W7Xlms05/GAM/104.719/gamdiagions_act.dat')
E28 = (data28[:,1]*(1/data28[0, 1]))
e28 = (data28[:,2]*(1/data28[0, 1]))
t28 = data28[:,0]

data29 = pylab.loadtxt('C:/Users/s162605/Documents/Studie/JAAR3/BEP/sterren/IFERC/W7Xlms05/GAM/136.837/gamdiagions_act.dat')
E29 = (data29[:,1]*(1/data29[0, 1]))
e29 = (data29[:,2]*(1/data29[0, 1]))
t29 = data29[:,0]

data39 = pylab.loadtxt('C:/Users/s162605/Documents/Studie/JAAR3/BEP/sterren/IFERC/W7Xlms09/GAM/104.719/gamdiagions_act.dat')
E39 = (data39[:,1]*(1/data39[0, 1]))
e39 = (data39[:,2]*(1/data39[0, 1]))
t39 = data39[:,0]

Q28 = np.sqrt(E28**2+e28**2)
Q29 = np.sqrt(E29**2+e29**2)
Q39 = np.sqrt(E39**2+e39**2)
```
